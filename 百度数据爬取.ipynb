```python
import requests
from lxml import etree
import urllib.parse
```

```python
url = 'https://baike.baidu.com/item/巴塞罗那足球俱乐部/2431429'
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36 Edg/89.0.774.57'}
r = requests.get(url,headers=headers)
#r=urllib.request.urlopen(url)
r.raise_for_status()
r.encoding = 'utf-8'
print("r.apparent_encoding", r.apparent_encoding)
print("r.text:",r.text)
```

```python
html = etree.HTML(r.text)
dt = html.xpath('/html/body/div[3]/div[2]/div/div[1]/div[7]//dt')
dd = html.xpath('/html/body/div[3]/div[2]/div/div[1]/div[7]//dd')
data_dict = {}
for i in range(len(dt)):
    # print(f'{i}{dt[i].xpath("string(.)")}')
    data_dict[''.join(dt[i].xpath("string(.)").split())] = ''.join(dd[i].xpath("string(.)").strip('\n').split())
print(data_dict)
```

## 批量获取所有球队的infobox

```python
teamsUrl = 'https://baike.baidu.com/item/%E8%A5%BF%E7%8F%AD%E7%89%99%E8%B6%B3%E7%90%83%E7%94%B2%E7%BA%A7%E8%81%94%E8%B5%9B/1470618?fr=aladdin'
teamresp = requests.get(teamsUrl, headers=headers)
teamresp.raise_for_status()
teamresp.encoding = 'utf-8'
teamHtml = etree.HTML(teamresp.text)
tr = teamHtml.xpath('/html/body//table[1]//tr')
ta = []
for i in range(int(len(tr)/2)):
    ta.extend(tr[1+2*i].xpath('.//a'))
teamList = []
for i in range(len(ta)):
    teamList.append((ta[i].xpath('.//text()')[0], ta[i].xpath('.//@href')[0]))
print(teamList)
```

```python
baseUrl = 'https://baike.baidu.com'
team_dict = {}
for i in range(len(teamList)):
    team_url = baseUrl + teamList[i][1]
    print(teamList[i], team_url)
    team_r = requests.get(team_url,headers=headers)
    team_r.raise_for_status()
    team_r.encoding = 'utf-8'
    team_html = etree.HTML(team_r.text)
    dt = team_html.xpath('/html/body/div[3]/div[2]/div/div[1]/div[7]//dt')
    dd = team_html.xpath('/html/body/div[3]/div[2]/div/div[1]/div[7]//dd')
    data_dict = {}
    print(i)
    if len(dt) == 0:
        print(f'This team has no infomation: {teamList[i][0]}')
    else:
        for j in range(len(dt)):
            data_dict[''.join(dt[j].xpath("string(.)").split())] = ''.join(dd[j].xpath("string(.)").strip('\n').split())
        team_dict[teamList[i][0]] = data_dict
print(team_dict)
```
